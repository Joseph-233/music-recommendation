```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
from sklearn.pipeline import Pipeline
```

```python
#MANUELA: ENCONTRA LA MANERA DE IMPORTAR ESTA FUNCION DIRECTO DEL OTRO DOC ASI NO HAY QUE COPY PASTE (MALA PRACTICA)

def classify_refined_heart_rate_intensity(row, max_hr):
    percentage_of_max = row['avg_heart_rate'] / max_hr
    if percentage_of_max <= 0.6:
        return 'Zone 1 (Very Light)'
    elif percentage_of_max <= 0.7:
        return 'Zone 2 (Light)'
    elif percentage_of_max <= 0.8:
        return 'Zone 3 (Moderate)'
    elif percentage_of_max <= 0.9:
        return 'Zone 4 (Hard)'
    else:
        return 'Zone 5 (Maximum)'
```

```python
df_running_data = pd.read_csv('../datasets/s1_laps_summary.csv')
df = pd.read_csv('../datasets/user_song_affinities_df.csv')
```

```python
# Find the maximum heart rate in the dataset to use as a proxy for 100% MHR
max_heart_rate = df_running_data['avg_heart_rate'].max()
```

```python
# Function to recommend unique songs based on user's heart rate
def recommend_unique_songs_based_on_heart_rate(user_heart_rate, df_running_data_new, df, max_heart_rate):
    # Determine the user's intensity zone
    user_intensity_zone = classify_refined_heart_rate_intensity({'avg_heart_rate': user_heart_rate}, max_heart_rate)

    # Filter the Spotify dataset for songs in the same intensity zone
    relevant_songs = df[df['intensity_zone'] == user_intensity_zone]

    # Rank these songs by popularity and remove duplicates
    ranked_songs = relevant_songs.sort_values(by='track_popularity', ascending=False).drop_duplicates(subset=['track_name'])

    # Recommend the top 5 unique songs
    top_5_unique_songs = ranked_songs.head(15)[['track_name', 'track_artist', 'track_popularity']]
    
    return top_5_unique_songs

# Get unique recommendations for the example user
unique_recommended_songs = recommend_unique_songs_based_on_heart_rate(200, df_running_data, df, max_heart_rate)
unique_recommended_songs
```
